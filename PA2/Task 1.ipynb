# Task 1: Working with a dataset with categorical features

```python
#IMPORTS
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction import DictVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler

```

### Step 1. Reading the data

```python
train_data = pd.read_csv('ADULT_TRAIN.csv')
test_data = pd.read_csv('ADULT_TEST.csv')
#train_data.head()
#train_data.columns
#train_data.shape
#test_data.shape
X_train_dict = train_data.iloc[:,:-1].to_dict('records')
print(X_train_dict[0])
Y_train = train_data.iloc[:,-1:]
X_test_dict = test_data.iloc[:,:-1].to_dict('records')
Y_test = test_data.iloc[:,-1:]

```

### Step 2.Encoding the features as numbers.

The Preprocessing technique DictVectorizer() is used .It  converts the feature values from a dictionary form into a matrix form to train a model with the inputs.

```python
#creating a dict vectorizer

vect = DictVectorizer()
train_values = vect.fit_transform(X_train_dict)
test_values = vect.transform(X_test_dict)
```

```python
grad_model = GradientBoostingClassifier()
grad_model.fit(train_values,Y_train)
grad_model.score(test_values,Y_test)
```

### Step 3. Combining the steps.

In this a model the Pipeline class is used to sequence the multiple process together in a single object .Here the Gradient boosting model is trained to using the pipeline class object. 

```python
grad_pipeline =make_pipeline(DictVectorizer(), GradientBoostingClassifier())
grad_pipeline.fit(X_train_dict,Y_train)
y_predict = grad_pipeline.predict(X_test_dict)
grad_pipeline.score(X_test_dict,Y_test)

```
